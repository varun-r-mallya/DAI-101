```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os



```

```python
df = pd.read_csv("Tesla Stock Price (2010 to 2023).csv")
```

```python
df.head()
```

```python
# dividing of data(column 'close') in sub categories
```

```python
df_low1=df.loc[df['Close']<80]#no. of entries(in column close) which are less than 80 come in this sub categrie
df_low2=df.loc[(df['Close']<160) & (df['Close']>=80)]#geater than equal to 80 and less than 160
df_low3=df.loc[(df['Close']<240) & (df['Close']>=160)]#geater than equal to 160 and less than 240
df_low4=df.loc[(df['Close']<320) & (df['Close']>=240)]#geater than equal to 240 and less than 320
df_high1=df.loc[(df['Close']<400) & (df['Close']>=320)]#geater than equal to 320 and less than 400
df_high2=df.loc[df['Close']>=400]#geater than equal 400
```

```python
#find no. of rows and columns
```

```python
df.shape
```

```python
## Univariate Analysis
```

```python
plt.plot(df_low1['Close'],np.zeros_like(df_low1['Close']), 'o') ## blue
plt.plot(df_low2['Close'],np.zeros_like(df_low2['Close']), 'o') ## orange
plt.plot(df_low3['Close'],np.zeros_like(df_low3['Close']), 'o') ## green
plt.plot(df_low4['Close'],np.zeros_like(df_low4['Close']), 'o') ## red
plt.plot(df_high1['Close'],np.zeros_like(df_high1['Close']), 'o') ## violet
plt.plot(df_high2['Close'],np.zeros_like(df_high2['Close']), 'o')
plt.xlabel('Closing Price')
plt.show()
```

```python
#count of data in each sub categorie
```

```python
print(df_low1.size)
print(df_low2.size)
print(df_low3.size)
print(df_low4.size)
print(df_high1.size)
print(df_high2.size)
print(df.size)
```

```python
lp1 = ((df_low1.size/df.size)*100)
lp2 = ((df_low2.size/df.size)*100)
lp3 = ((df_low3.size/df.size)*100)
lp4 = ((df_low4.size/df.size)*100)
hp1 = ((df_high1.size/df.size)*100)
hp2 = ((df_high2.size/df.size)*100)
```

```python
#pie chart 
```

```python
pie_data = pd.DataFrame({'Name': ['low1', 'low2', 'low3',
                                   'low4', 'high1', 'high2'],
                          'percent of each categ': [df_low1.size, df_low2.size, df_low3.size,
                                                  df_low4.size, df_high1.size, df_high2.size]})

pie_data.groupby(['Name']).sum().plot(kind='pie', y='percent of each categ', autopct='%2.2f')
  
```

```python
#bar graph
pie_data.groupby(['Name']).sum()
```
