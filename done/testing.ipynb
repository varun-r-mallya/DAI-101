```python
class balls:
    def __init__(self, color, size):
        self.color = color
        self.size = size
    def __str__(self):
        return f'{self.color} ball of size {self.size}'
    def __iter__(self):
        return iter([self.color, self.size])
    def __next__(self):
        return next([self.color, self.size])
    
    
b = balls('red', 5)
mtitr = iter(b)
print(mtitr)
print(next(mtitr))
print(*b)
str(b)
```
